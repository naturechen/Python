```python
def getData(companyCode):
    
   
    api_result = requests.get('https://ies-midterm.soulution.rocks/data/'+companyCode)
    
    api_response = api_result.json()
    
    return api_response
    
HIKData = getData('557b58dd1d')

getHIKData = pd.DataFrame(HIKData['data']['data'])

```
